```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from pprint import pprint
```

```python
#path_file = Path to folder with text files
folder_texts = '/Users/anzalks/Documents/git_sync/camp_data/'
```

```python
def list_files(p):
    f_list = []
    p = Path(p)
    freq=list(p.glob('**/frequency*'))
    freq = sorted(freq)
    cond=list(sorted(p.glob('**/gIH*')))
    freq = freq[::4]
    #pprint(f' sorted = {freq}')
    cond.sort()
    return freq, cond

def frq_wt(text_file):
    frq_list = []
    frq_range = []
    for f in text_file:
        t_f = np.loadtxt(f)
        t_f=np.squeeze(t_f)
        t_f=np.transpose(t_f)
        #print(t_f)
        f = str(f.stem).split('_')[-1]
        frq_list.append(t_f)
        frq_range.append(f)
    #
    #print(f'squeeze = {frq_list[0:5]}')
    #frq_list = np.transpose(frq_list)
    #print(f'transpose={frq_list[0:5]}')
    #print(frq_list[1])
    #print(f' f list ={f}')
    return frq_range, frq_list

def frq_wt(text_file):
    frq_list = []
    frq_range = []
    for f in text_file:
        t_f = np.loadtxt(f)
        t_f=np.squeeze(t_f)
        t_f=np.transpose(t_f)
        #print(t_f)
        f = str(f.stem).split('_')[-1]
        frq_list.append(t_f)
        frq_range.append(f)
    #
    #print(f'squeeze = {frq_list[0:5]}')
    #frq_list = np.transpose(frq_list)
    #print(f'transpose={frq_list[0:5]}')
    #print(frq_list[1])
    #print(f' f list ={f}')
    return frq_range, frq_list



def plot_all_freq(frq_range, frq_list):
    plt.figure()
    for i,rn in enumerate(frq_range):
        plt.plot(frq_list[i][0]/1000,frq_list[i][1],label=(f'frequency = {rn} Hz'))
    plt.xlabel('Time (s)')
    plt.ylabel('Synaptic weight')
    plt.title('Synaptic weight change with time and frequency')
    plt.xlim([0,30])
    plt.legend(bbox_to_anchor =(1.1, 0.5))
    plt.show()
    
def plot_all_cond(cond_range, cond_freq):
    plt.figure()
    for i,rn in enumerate(cond_range):
        plt.plot(cond_freq[i][0],cond_freq[i][1],label=(f'conductance = {rn} S/cm^2'))
    plt.xlabel('Frequency')
    plt.ylabel('% change in synaptic weight')
    plt.title('Sliding threshold with conductance change')
    #plt.xlim([0,30])
    plt.legend(bbox_to_anchor =(1.8, 0.5))
    plt.show()
    
```

```python
freq_f,cond_f = list_files(folder_texts)
frq_range, frq_list = frq_wt(freq_f)
cond_range, cond_freq = frq_wt(cond_f)
```

```python
plot_all_freq(frq_range, frq_list)
```

```python
plot_all_cond(cond_range, cond_freq)
```
